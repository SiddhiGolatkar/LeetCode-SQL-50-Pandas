```python
import pandas as pd
import datetime as dt

def rising_temperature(weather: pd.DataFrame) -> pd.DataFrame:
    weather['recordDate'] = pd.to_datetime(weather['recordDate'])
    weather.sort_values('recordDate', inplace = True)
    weather.reset_index(inplace= True)
    temp = weather['temperature'][1:]
    temp = temp.reset_index()
    temp2 = weather['recordDate'][1:]
    temp2 = temp2.reset_index()
    temp3 = weather['id'][1:]
    temp3 = temp3.reset_index()
    weather['Id'] = temp3['id']
    weather['next_day_temp'] = temp['temperature']
    weather['next_date'] = temp2['recordDate']
    weather['next_day_temp'].fillna(0, inplace= True)
    weather['Id'] = weather['Id'].fillna(0)
    weather['greater'] = weather['next_day_temp'] > weather['temperature']
    weather['next_date'].fillna(weather['recordDate'] + pd.Timedelta(days=1), inplace = True)
    weather['no_of_days'] = (weather['next_date'] - weather['recordDate']).dt.days
    result = weather[(weather['greater'] == True) & (weather['no_of_days'] == 1) & (weather['Id'] > 0)]['Id']
    result = result.reset_index()
    return result[['Id']]
```

```python
def rising_temperature(weather: pd.DataFrame) -> pd.DataFrame:
    weather.sort_values(by='recordDate', inplace=True)
    return weather[
        (weather.temperature.diff() > 0)
      & (weather.recordDate.diff().dt.days == 1)
    ][['id']]
```

```python
def get_average_time(activity: pd.DataFrame) -> pd.DataFrame:
    
    activity['machine_id'] = activity['machine_id'].astype(str)
    activity['process_id'] = activity['process_id'].astype(str)
    activity['id'] = activity['machine_id'] + activity['process_id']

    start = activity[activity['activity_type'] == 'start']
    start.rename(columns = {'timestamp' : 'start_time'}, inplace = True)

    end = activity[activity['activity_type'] == 'end']
    end.rename(columns = {'timestamp' : 'end_time'}, inplace = True)

    end = end[['end_time', 'id']]

    process = start.merge(end, on = 'id')
    process['machine_id'] = process['machine_id'].astype('int64')
    process['processing_time'] = process['end_time'] - process['start_time']

    return round(process.groupby('machine_id')[['processing_time']].mean(), 3).reset_index()
```

Each row of this table indicates that the user with ID user_id requested a confirmation message at time_stamp and that confirmation message was either confirmed ('confirmed') or expired without confirming ('timeout').
 
The confirmation rate of a user is the number of 'confirmed' messages divided by the total number of requested confirmation messages. The confirmation rate of a user that did not request any confirmation messages is 0. Round the confirmation rate to two decimal places.

Write a solution to find the confirmation rate of each user.

```python
import pandas as pd

def confirmation_rate(signups: pd.DataFrame, confirmations: pd.DataFrame) -> pd.DataFrame:
    action = confirmations['action'].unique()
    if 'confirmed' in action:
    
        result = confirmations.groupby('user_id')[['action']].count().reset_index()

        a = confirmations.groupby(['user_id', 'action'])[['action']].count().unstack().fillna(0)

        a.reset_index(inplace = True)

        result['confirmed'] = a['action']['confirmed'].values
        result['timeout'] = a['action']['timeout'].values

        result['confirmation_rate'] = np.where(result['confirmed'], result['confirmed'] / result['action'], 0)

        result['confirmation_rate'] = round(result['confirmation_rate'], 2)

        result = result.merge(signups, on = 'user_id', how = 'right')
   
        result.fillna(0.00, inplace = True)

        return result[['user_id', 'confirmation_rate']]

    else:
        signups['confirmation_rate'] = 0.00
        return signups[['user_id', 'confirmation_rate']]
```

```python
import pandas as pd

def users_percentage(users: pd.DataFrame, register: pd.DataFrame) -> pd.DataFrame:
    result = register.groupby('contest_id')[['user_id']].count().reset_index().rename(columns={'user_id' : 'count_of_users'})

    result['percentage'] = round((result['count_of_users'] / users['user_id'].nunique()) * 100, 2)

    result = result[['contest_id', 'percentage']].sort_values(['percentage', 'contest_id'], ascending = [False, True])

    return result
```

```python
users_percentage(users, register)
```
